```python
import torch
from transformers import BertForMaskedLM, BertTokenizer

model_name = 'bert-base-uncased'
# 1 load the model from bert
model = BertForMaskedLM.from_pretrained(model_name)
# 2 create a tokeniser from huggingface
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
# [MASK] depicts the word to complete

sentence = "The quick brown [MASK] jumps over the lazy dog"

# 3. tokenise the text
tokenised_text = tokenizer.tokenize(sentence)

# 4. replace token of [MASK] with [MASK]
masked_index = tokenised_text.index('[MASK]')
tokenised_text[masked_index] = '[MASK]'

print(tokenised_text)
print(masked_index)

# 5. convert tokens to embeddings
indexed_tokens = tokenizer.convert_tokens_to_ids(tokenised_text)
tokens_tensor = torch.tensor([indexed_tokens])

print(indexed_tokens)
print(tokens_tensor)
```

```python
# 6. pass the embeddings to the model make prediction
with torch.no_grad():
  outputs = model(tokens_tensor)
  # 7. get the top 5 most likely token ID
  predictions = outputs[0][0, masked_index].topk(10).indices.tolist()
  print(predictions)

  # 8. convert token ID to corresponding text
  predicted_tokens = tokenizer.convert_ids_to_tokens(predictions)
  print(predicted_tokens)
```
